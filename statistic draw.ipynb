```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use ('ggplot')

x = [1, 2, 3, 4, 5]
y = [12, 23, 43, 12, 65]
y_err = [1, 50, 3, 4, 5]
x_err = [0.4]*5


plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('rainfal')
plt.title('rain over the week',color ="blue")
plt.suptitle('supertitle',y=1)
plt.errorbar(x, y,xerr = x_err,yerr = y_err, ecolor = 'black')
plt.savefig('figure 1.svg')


```
